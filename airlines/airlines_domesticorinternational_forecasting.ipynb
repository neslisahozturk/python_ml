```python
import pandas as pnd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
airlines_data=pnd.read_csv('cleaned_air.csv')
```

```python
airlines_data.shape
```

```python
airlines_data.head()
```

```python
X=airlines_data.drop(columns=['Operating Airline IATA Code','Published Airline IATA Code','GEO Summary','Aircraft Manufacturer','Aircraft Model'],axis=1)
lbe=LabelEncoder()
y=lbe.fit_transform(airlines_data['GEO Summary'])
```

```python
categorical_columns = X.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(X[categorical_columns])
one_hot_X = pnd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
X_encoded = pnd.concat([X, one_hot_X], axis=1)
X_encoded = X_encoded.drop(categorical_columns, axis=1)
```

```python
X_encoded
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X_encoded)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
training_data_accuracy
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,y_test)
test_data_accuracy
```
